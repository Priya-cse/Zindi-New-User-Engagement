### KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init='k-means++', n_clusters=2, n_init=2)
```

```python
kmeans.fit(df_temp[nameList])
```

```python
y_pred = kmeans.predict(df_temp[nameList])
```

```python
y_pred
```

```python
op.countOf(y_pred, 0)
```

```python
for i in range(len(y_pred)):
    if y_pred[i] == 0:
        y_pred[i] = 1
    elif y_pred[i] == 1:
        y_pred[i] = 0
#     elif y_pred[i] == 1:
#         y_pred[i] = 0
```

```python
op.countOf(y_pred, 0)
```

```python
plt.scatter(X_proj[:,0], X_proj[:,1], c=y_pred)
plt.title("KMeans Clustering Results")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(["1", "0"], loc ="upper left")
```

```python
plt.scatter(df_temp["count_UA"], df_temp["Day_duration"], c=y_pred)
```

```python
plt.scatter(df_temp["Successful_Submission_Count"], df_temp["count_UA"], c=y_pred)
```

```python
plt.scatter(df_temp["Day_duration"], df_temp["Day_last_active"], c=y_pred)
```
