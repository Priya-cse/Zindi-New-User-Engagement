```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
def remove_id(x):
    x = x.replace('$', '')
    if 'ID' in x:
        return "_".join(x.split('_')[:-1])
    
    return x.replace(' ', '_')

df_userAct.Title.apply(remove_id).value_counts()
```

```python
df_userAct['main_activities'] = df_userAct.Title.apply(remove_id)
df_userAct.main_activities.unique()
```

```python
act_vc = df_userAct.main_activities.value_counts()
selected_activities = act_vc[act_vc > 1000].keys()
selected_activities
```

```python
df_userAct = df_userAct[df_userAct.main_activities.isin(selected_activities)].reset_index(drop=True)
df_userAct
```

```python
dataset = df_userAct.groupby(['User_ID', 'main_activities']).size().reset_index().\
    rename(columns={0: 'count'}).\
        sort_values(['User_ID', 'count'], ascending=False)

dataset 
```

```python
dataset.main_activities.nunique()
```

```python
all_activities = dataset.main_activities.unique()
```

```python
dataset[dataset.main_activities == "comp_ID"].drop(columns='main_activities').reset_index(drop=True)
```

```python
for activity in all_activities:
    dataset[activity] = np.zeros((dataset.shape[0], 1))
```

```python
dataset
```

```python
for i in range(len(dataset)):
    for activity in all_activities:
        if dataset.loc[i, "main_activities"] == activity:
            dataset.loc[i, activity] = dataset.loc[i, "count"] 
```

```python
dataset = dataset.groupby("User_ID").sum().reset_index()
```

```python
dataset.drop("count", inplace=True, axis=1)
```

```python
dataset.info()
```

```python
dataset
```

```python
df_userAct = pd.merge(df_userAct, dataset, how="left", on=['User_ID'])
```

```python
df_userAct.drop("Title", inplace=True, axis=1)
df_userAct.drop("main_activities", inplace=True, axis=1)
```

```python
df_userAct.rename(columns={'Theme': 'Blog Theme'},  inplace = True)
```

```python
df_userAct.info()
```

```python
df_temp2 = df_userAct.groupby('User_ID').mean() 
```

```python
df_temp3 = df_userAct.groupby('User_ID').count() 
```

```python
df_temp4 = df_userAct.groupby('User_ID').min() 
```

```python
df_temp = df_temp2
```

```python
#Last day active
df_temp["datetime Day_of_month"] = df_temp1["datetime Day_of_month"]
```

```python
df_temp["count_UA"] = df_temp3["datetime Day_of_month"]
```

```python
df_temp["Day duration"] = df_temp1["datetime Day_of_month"] - df_temp4["datetime Day_of_month"]
```
