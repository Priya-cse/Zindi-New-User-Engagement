## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
import operator as op
```

```python
df_temp = df_train.copy()
```

```python
y = df_temp["Active"]
```

```python
my_cols = set(df_temp.columns)
my_cols.remove('Active')
X = df_temp[my_cols]
```

```python
for column in X:
    X[column] = X[column] / X[column].abs().max()
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=12, shuffle=True)
```

```python
clf = LogisticRegression(random_state=0).fit(train_X, train_y)
```

```python
y_p = clf.predict(test_X)
```

```python
op.countOf(y_p, 1)
```

```python
accuracy_score(test_y, y_p)
```

```python
f1_score(test_y, y_p)
```
